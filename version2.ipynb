# Filtrage ile-De-France


```python

# Afficher les noms de colonnes
print(df2.columns)

# Filtrer les lignes où la colonne "Région" est "Ile-de-France"
idf_df2 = df2[df2['Région'] == 'Ile-de-France']

# Afficher le DataFrame résultant
print(idf_df2)
```

```python
# Fonction pour obtenir le code INSEE à partir du code postal pour la région Île-de-France
def get_code_insee_idf(code_postal):
    if code_postal.startswith('75'):  # Paris
        arrondissement = int(code_postal[3:])
        if 1 <= arrondissement <= 20:
            return f'751{arrondissement:02d}'
    elif code_postal.startswith('77'):  # Seine-et-Marne
        return '770'  # Exemple de code INSEE pour Seine-et-Marne
    elif code_postal.startswith('78'):  # Yvelines
        return '780'  # Exemple de code INSEE pour Yvelines
    elif code_postal.startswith('91'):  # Essonne
        return '910'  # Exemple de code INSEE pour Essonne
    elif code_postal.startswith('92'):  # Hauts-de-Seine
        return '920'  # Exemple de code INSEE pour Hauts-de-Seine
    elif code_postal.startswith('93'):  # Seine-Saint-Denis
        return '930'  # Exemple de code INSEE pour Seine-Saint-Denis
    elif code_postal.startswith('94'):  # Val-de-Marne
        return '940'  # Exemple de code INSEE pour Val-de-Marne
    elif code_postal.startswith('95'):  # Val-d'Oise
        return '950'  # Exemple de code INSEE pour Val-d'Oise
    # Ajoutez d'autres correspondances au besoin pour d'autres départements de la région Île-de-France
    else:
        return None

# Appliquer la fonction pour créer une nouvelle colonne "Code INSEE"
idf_df2['Code INSEE'] = idf_df2['Code postal'].apply(lambda x: get_code_insee_idf(x))

# Afficher le DataFrame résultant avec la nouvelle colonne
print(idf_df2)
```
